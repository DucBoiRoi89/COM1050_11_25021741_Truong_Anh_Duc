```python
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h
    def s(self):
        self.total = round(2 * 3.14 * self.r * self.h, 2)
        return self.total
    def v(self):
        self.sum = round(3.14 * (self.r ** 2) * self.h, 2)
        return self.sum
        
r, h = map(float, input().split())
p = Cylinder(r, h)
print(p.s(), p.v())
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = []
    def add_item(self, name, price):
        self.items.append({"name": name, "price": price})
    def remove_item(self, name):
        for i in range(len(self.items)):
            if self.items[i]["name"] == name:
                del self.items[i]
                print(f"Đã xóa '{name}' khỏi giỏ hàng.")
                return
        print(f"Không tìm thấy sản phẩm '{name}'.")
    def is_empty(self):
        return len(self.items) == 0
    def get_total(self):
        return sum(item["price"] for item in self.items)
    def display_cart(self):
        if self.is_empty():
            print("Giỏ hàng đang rỗng.")
        else:
            for item in self.items:
                print(f"- {item['name']}: {item['price']}")
    def clear_cart(self):
        self.items = []
cart = ShoppingCart()
while True:
    choice = input()
    if choice == '2':
        name = input()
        price = float(input())
        cart.add_item(name, price)
    elif choice == '3':
        name = input()
        cart.remove_item(name)
    elif choice == '4':
        print("Giỏ hàng rỗng." if cart.is_empty() else "Giỏ hàng không rỗng.")
    elif choice == '5':
        print(f"{cart.get_total()}")
    elif choice == '6':
        cart.display_cart()
    elif choice == '7':
        cart.clear_cart()
    exit_choice = input("Có muốn thoát khỏi chương trình không?")
    if exit_choice.lower() == 'yes':
        break
```

```python
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h
    def perimeter(self):
        return 2 * (self.w + self.h)
    def scale(self, k):
        self.w *= k
        self.h *= k
try:
    data = input().split()
    if len(data) == 3:
        w, h, k = map(float, data)
        rect = Rectangle(w, h)
        rect.scale(k)
        print(f"{rect.area():.2f} {rect.perimeter():.2f}")
except (EOFError, ValueError):
    pass
```

```python
import math
class Fraction:
    def __init__(self, num, den):
        if den == 0: raise ValueError("Mẫu số phải khác 0")
        common = math.gcd(num, den)
        self.num = num // common
        self.den = den // common
        if self.den < 0:
            self.num = -self.num
            self.den = -self.den
    def add(self, other):
        return Fraction(self.num * other.den + other.num * self.den, self.den * other.den)
    def sub(self, other):
        return Fraction(self.num * other.den - other.num * self.den, self.den * other.den)
    def mul(self, other):
        return Fraction(self.num * other.num, self.den * other.den)
    def div(self, other):
        return Fraction(self.num * other.den, self.den * other.num)
    def __str__(self):
        if self.den == 1: return str(self.num)
        return f"{self.num}/{self.den}"
try:
    line = input().split()
    if len(line) == 5:
        a, b, op, c, d = int(line[0]), int(line[1]), line[2], int(line[3]), int(line[4])
        f1, f2 = Fraction(a, b), Fraction(c, d)
        if op == '+': res = f1.add(f2)
        elif op == '-': res = f1.sub(f2)
        elif op == '*': res = f1.mul(f2)
        elif op == '/': res = f1.div(f2)
        print(res)
except Exception: pass
```

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)
    def deposit(self, x):
        self.balance += float(x)
    def withdraw(self, x):
        amount = float(x)
        if amount <= self.balance:
            self.balance -= amount
    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"
try:
    line1 = input().split()
    owner, initial_balance = line1[0], float(line1[1])
    account = BankAccount(owner, initial_balance)
    n = int(input().strip())
    for _ in range(n):
        command = input().split()
        type, x = command[0], float(command[1])
        if type == "DEPOSIT": account.deposit(x)
        elif type == "WITHDRAW": account.withdraw(x)
    print(account)
except (EOFError, ValueError, IndexError): pass
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []
    def add_score(self, subject, score):
        self.scores.append(float(score))
    def avg(self):
        return sum(self.scores) / len(self.scores) if self.scores else 0
    def rank(self):
        a = self.avg()
        if a >= 8: return "Excellent"
        if a >= 6.5: return "Good"
        if a >= 5: return "Average"
        return "Poor"
    def __str__(self):
        a = self.avg()
        return f"{self.name} {a:.2f} {self.rank()}"
try:
    name = input().strip()
    n = int(input().strip())
    student = Student(name)
    for _ in range(n):
        line = input().split()
        if len(line) >= 2:
            subject, score = line[0], line[1]
            student.add_score(subject, score)
    print(student)
except (EOFError, ValueError): pass
```

```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = int(year)
class Library:
    def __init__(self):
        self.books = []
    def add(self, book):
        self.books.append(book)
    def count_by_author(self, name):
        return sum(1 for b in self.books if b.author.lower() == name.lower())
    def find_by_year(self, y):
        return sum(1 for b in self.books if b.year == int(y))
try:
    n_input = input().strip()
    if n_input:
        n = int(n_input)
        lib = Library()
        for _ in range(n):
            line = input().strip()
            if line.startswith("ADD"):
                parts = line[4:].split(";")
                if len(parts) == 3:
                    lib.add(Book(parts[0], parts[1], parts[2]))
            elif line.startswith("COUNTYEAR"):
                year = line.split()[-1]
                print(lib.find_by_year(year))
            elif line.startswith("COUNT"):
                author = line[6:].strip()
                print(lib.count_by_author(author))
except (EOFError, ValueError): pass
```

```python
import math
class Vector:
    def __init__(self, x, y, z):
        self.x, self.y, self.z = x, y, z
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y, self.z - other.z)
    def dot(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z
    def cross(self, other):
        return Vector(self.y * other.z - self.z * other.y, self.z * other.x - self.x * other.z, self.x * other.y - self.y * other.x)
    def length(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)
def solve():
    try:
        points = []
        for _ in range(4):
            points.append(Vector(*map(float, input().split())))
        a, b, c, d = points
        ab, bc, cd = b - a, c - b, d - c
        x = ab.cross(bc)
        y = bc.cross(cd)
        cos_phi = x.dot(y) / (x.length() * y.length())
        cos_phi = max(-1.0, min(1.0, cos_phi))
        phi = math.degrees(math.acos(cos_phi))
        print(f"{phi:.2f}")
    except (EOFError, ValueError, ZeroDivisionError): pass
if __name__ == "__main__":
    solve()
```

```python
class ComplexNumber:
    def __init__(self, initRealPart, initImaginaryPart):
        self.realPart = float(initRealPart)
        self.imaginaryPart = float(initImaginaryPart)
    def multiply(self, other):
        new_real = self.realPart * other.realPart - self.imaginaryPart * other.imaginaryPart
        new_imaginary = self.realPart * other.imaginaryPart + self.imaginaryPart * other.realPart
        return ComplexNumber(new_real, new_imaginary)
    def __str__(self):
        return f"{int(self.realPart)} {int(self.imaginaryPart)}"
try:
    r1, i1 = map(float, input().split())
    r2, i2 = map(float, input().split())
    c1 = ComplexNumber(r1, i1)
    c2 = ComplexNumber(r2, i2)
    result = c1.multiply(c2)
    print(result)
except (EOFError, ValueError): pass
```

```python
import math
class Triangle:
    def __init__(self, initEdge1, initEdge2, initEdge3):
        self.edge1, self.edge2, self.edge3 = initEdge1, initEdge2, initEdge3
    def getArea(self):
        a, b, c = self.edge1, self.edge2, self.edge3
        if a + b > c and a + c > b and b + c > a and a > 0 and b > 0 and c > 0:
            s = (a + b + c) / 2.0
            area = math.sqrt(s * (s - a) * (s - b) * (s - c))
            return area
        return -1
try:
    data = input().split()
    if len(data) == 3:
        e1, e2, e3 = map(int, data)
        tri = Triangle(e1, e2, e3)
        res = tri.getArea()
        if res == -1: print("invalid")
        else:
            if res == int(res): print(int(res))
            else: print(round(res, 2))
except (EOFError, ValueError): pass
```
