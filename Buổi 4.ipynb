```python
import math
def prime_number_testing():
    while True:
        n = int(input())
        if n <= 0:
            print("Lỗi. Vui lòng nhập lại giá trị n.")
        else:
            if n < 2:
                print("False")
                return False
            else:
                for i in range (2, math.floor(math.sqrt(n)) + 1):
                    if n % i == 0:
                        print("False")
                        return False
                print("True")
                return True
            break
prime_number_testing()
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def sum_primes_in_range():
    while True:
        try:
            a, b = map(int, input().split())
            total = 0
            for num in range(a, b + 1):
                if is_prime(num):
                    total += num
            print(total)
        except EOFError:
            break
sum_primes_in_range()
```

```python
def largest_prime_factor():
    n = int(input())
    factor = 2
    largest = 1
    while factor * factor <= n:
        if n % factor == 0:
            largest = factor
            n //= factor
        else:
            factor += 1
    if n > 1:
        largest = n
    print(int(largest))
largest_prime_factor()
```

```python
def reverse_number(n):
    return int(str(n)[::-1])
def is_palindrome(n):
    return str(n) == str(n)[::-1]
def reverse_add_palindrome():
    n = int(input())
    steps = 0
    while not is_palindrome(n):
        n = n + reverse_number(n)
        steps += 1
    print(steps, n)
reverse_add_palindrome()
```

```python
def perfect_squares_distinct_digits():
    n = int(input())
    found = False
    for i in range(1, int(n**0.5) + 1):
        sq = i * i
        if sq <= n:
            s = str(sq)
            if len(s) == len(set(s)):  # không trùng chữ số
                print(sq, end=" ")
                found = True
    if not found:
        print("no number")
perfect_squares_distinct_digits()
```

Bài 10

```python
def collatz_length(n):
    count = 1
    while n != 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = 3 * n + 1
        count += 1
    return count

def longest_collatz_up_to_n():
    n = int(input())
    max_len = 0
    result = 1
    for i in range(1, n + 1):
        length = collatz_length(i)
        if length > max_len:
            max_len = length
            result = i
    print(result, max_len)

longest_collatz_up_to_n()
```

```python
def sum_divisors(n):
    total = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            total += i
            if i != n // i:
                total += n // i
    return total
def close_pair_number():
    a, b = map(int, input().split())
    if sum_divisors(a) == b and sum_divisors(b) == a:
        print("true")
    else:
        print("false")
close_pair_number()
```

```python
def fibonacci_max(a):
    fib1, fib2 = 1, 1
    while fib2 <= a:
        fib1, fib2 = fib2, fib1 + fib2
    return fib1
A = int(input("Nhập số nguyên dương A: "))
print("Số Fibonacci lớn nhất không vượt quá A là:", fibonacci_max(A))
```

```python
def count_words(s):
    count = 0
    in_word = False
    for i in range(len(s)):
        if s[i] != ' ':
            if not in_word:
                count += 1
                in_word = True
        else:
            in_word = False
    return count
```

```python
def is_valid_password(s):
    if not (8 <= len(s) <= 16):
        return False
    has_special = False
    has_lower = False
    has_upper = False
    has_digit = False
    for ch in s:
        if 'a' <= ch <= 'z':
            has_lower = True
        elif 'A' <= ch <= 'Z':
            has_upper = True
        elif '0' <= ch <= '9':
            has_digit = True
        else:
            has_special = True

    return has_special and has_lower and has_upper and has_digit
```
