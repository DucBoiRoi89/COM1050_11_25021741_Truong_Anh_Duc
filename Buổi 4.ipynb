```python
import math
def prime_number_testing():
    while True:
        n = int(input())
        if n <= 0:
            print("Lỗi. Vui lòng nhập lại giá trị n.")
        else:
            if n < 2:
                print("False")
                return False
            else:
                for i in range (2, math.floor(math.sqrt(n)) + 1):
                    if n % i == 0:
                        print("False")
                        return False
                print("True")
                return True
            break
prime_number_testing()
```

```python
def reverse_n(n):
    return str(n)[::-1]
def palindrome(n):
    return str(n) == str(n)[::-1]
def reverse_add_palindrome():
    number_of_steps = 0
    n = int(input())
    while not palindrome(n):
        n = n + int(reverse_n(n))
        number_of_steps += 1
    print(number_of_steps, n)
reverse_add_palindrome()
```

Bài 10:

```python
def collatz_length(n):
    count = 1
    while n != 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = 3 * n + 1
        count += 1
    return count

def longest_collatz_up_to_n():
    n = int(input())
    max_len = 0
    result = 1
    for i in range(1, n + 1):
        length = collatz_length(i)
        if length > max_len:
            max_len = length
            result = i
    print(result, max_len)

longest_collatz_up_to_n()
```
