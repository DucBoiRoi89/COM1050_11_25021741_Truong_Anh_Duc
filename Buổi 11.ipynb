W8A3:

```python
def is_happy(n):
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(digit)**2 for digit in str(n))
    return n == 1
try:
    a = int(input())
    print("YES" if is_happy(a) else "NO")
except ValueError:
    print("Loi nhap lieu")
```

W8A5:

```python
def convert_to_int(my_list):
    res = []
    for x in my_list:
        try:
            res.append(int(x))
        except ValueError:
            print(f"Canh bao: '{x}' khong the chuyen doi")
    return res
my_list = list(map(int, input().split()))
print(convert_to_int(my_list))
```

```python
try:
    line = input().split()
    if not line: raise Exception
    n = int(line[0])
    arr = list(map(int, line[1:]))
    if len(arr) == n and len(set(arr)) == n and all(x > 0 for x in arr):
        print(sum(arr))
    else:
        print("Mang khong hop le")
except:
    print("Mang khong hop le")
```
