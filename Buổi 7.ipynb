```python
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (left + right) // 2
        if target == arr[mid]:
            return mid
        elif target > arr[mid]:
            left = mid + 1
        else:
            right = mid - 1
    return "-1"
arr = list(map(int, input().split()))
target = int(input())
left , right = 0, len(arr) - 1
print(binary_search(arr,left, right, target))
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        b = False
        for x in range(n - i - 1):
            if arr[x] > arr[x + 1]:
                arr[x], arr[x + 1] = arr[x + 1], arr[x]
                b = True
        if not b:
            break  
    return arr      
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        b = i
        for j in range(i + 1, n):
            if arr[j] < arr[b]:
                b = j
        arr[i], arr[b] = arr[b], arr[i] 
    return arr
arr = list(map(int, input().split()))
print(bubble_sort(arr))
print(selection_sort(arr))            
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        self.left_count = 0  
        self.dup = 1        
class Solution:
    def countSmaller(self, nums):
        if not nums:
            return []
        n = len(nums)
        res = [0] * n
        root = None
        for i in range(n - 1, -1, -1):
            root, count = self.insert(root, nums[i])
            res[i] = count        
        return res
    def insert(self, node, val):  
        if not node:
            return Node(val), 0
        if node.val == val:
            node.dup += 1
            return node, node.left_count   
        elif val < node.val:
            node.left_count += 1
            node.left, count = self.insert(node.left, val)
            return node, count    
        else:
            node.right, count = self.insert(node.right, val)         
            return node, node.left_count + node.dup + count   
sol = Solution()
s = input("Nhập danh sách: ")
clean_s = s.replace('[', '').replace(']', '').replace(',', ' ')
try:
    val = [int(x) for x in clean_s.split()]
    print(sol.countSmaller(val))
except ValueError:
    print("Lỗi: Vui lòng nhập đúng định dạng danh sách số!")
```

```python
import bisect

def maxSubarrayModulo(nums, m):
    prefix_sums = [0]
    curr_sum = 0
    max_mod = 0
    for x in nums:
        curr_sum = (curr_sum + x) % m
        idx = bisect.bisect_right(prefix_sums, curr_sum)
        if idx < len(prefix_sums):
            max_mod = max(max_mod, (curr_sum - prefix_sums[idx] + m) % m)
        max_mod = max(max_mod, curr_sum)
        bisect.insort(prefix_sums, curr_sum)
    return max_mod
try:
    raw_nums = input("Nhập danh sách nums (ví dụ [3, 3, 9, 9, 5]): ")
    m = int(input("Nhập số nguyên m: "))
    clean_nums = raw_nums.replace('[', '').replace(']', '').replace(',', ' ')
    nums = [int(x) for x in clean_nums.split()]
    print(maxSubarrayModulo(nums, m))
except ValueError:
    print("Vui lòng nhập đúng định dạng!")
```

```python
def fastest_finger_first():
    options = []
    labels = ['A', 'B', 'C', 'D']
    for i in range(4):
        phrase = input(f"{labels[i]}: ").strip()
        options.append((phrase, labels[i]))
    options.sort(key=lambda x: x[0])
    result = [item[1] for item in options]
    print(" ".join(result))
if __name__ == "__main__":
    fastest_finger_first()
```

```python
from collections import Counter
def find_longest_harmonious_subsequence(arr):
    counts = Counter(arr)
    max_length = 0
    for x in counts:
        if x + 1 in counts:
            max_length = max(max_length, counts[x] + counts[x + 1])
    return max_length
try:
    raw_input = input().strip()
    clean_input = raw_input.replace('[', '').replace(']', '').replace(',', ' ')
    nums = [int(x) for x in clean_input.split()]
    print(find_longest_harmonious_subsequence(nums))
except EOFError:
    pass
```

```python
def find_neighbors():
    try:
        n_input = input().strip()
        if not n_input: return
        n = int(n_input)
        guests = []
        for _ in range(n):
            guests.append(input().strip())
        if "Nemo" in guests:
            idx = guests.index("Nemo")
            name1 = guests[(idx - 1) % n]
            name2 = guests[(idx + 1) % n]
            print(f"{name1} and {name2}")
    except (EOFError, ValueError):
        pass
if __name__ == "__main__":
    find_neighbors()
```
