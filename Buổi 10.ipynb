```python
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(x, y)}
    for move in moves:
        if move == 'U':
            y += 1
        elif move == 'D':
            y -= 1
        elif move == 'R':
            x += 1
        elif move == 'L':
            x -= 1
        current_pos = (x, y)
        if current_pos in visited:
            return True
        visited.add(current_pos)
    return False
moves = input()
print(is_path_crossing(moves))
```

```python
def Median(nums):
    a, b, c, d, e = nums
    if a > b: a, b = b, a
    if c > d: c, d = d, c
    if a < c:
        a, b = c, d
        c, d = a, b
    if a < c:
        a, b = c, d
    if e < b:
        if e < a:
            if c < a: return a
            else: return c
        else:
            if c < e: return e
            else: return c
    else:
        if b < c: return c
        else: return b
nums = map(int, input().split())
print(Median(nums))
```

```python
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    res = 0
    i = 0
    for house in houses:
        while i < len(heaters) - 1 and abs(heaters[i+1] - house) <= abs(heaters[i] - house):
            i += 1
        res = max(res, abs(heaters[i] - house))
    return res
houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
print(findRadius(houses, heaters))
```
