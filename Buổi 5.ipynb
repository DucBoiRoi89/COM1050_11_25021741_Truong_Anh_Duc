```python
def so_nguyen_to():
    n = int(input())
    if n < 2:
        print("False")
        return False
    else:
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                print("False")
                return False
        print("True")
        return True
so_nguyen_to()
    
```

```python
def so_hoan_hao(n):
    ds = []
    for i in range(1, n):
        if n % i == 0:
            ds.append(i)
    total = 0
    for i in ds:
        total += i
    if total == n:
        print("True")
        return True
    else:
        print("False")
        return False
n = int(input())
so_hoan_hao(n)
```

```python
def a(arr, k):
    ds = []
    for i in range(len(arr)):
        if arr[i] == k:
            ds.append(i)
    if len(ds) > 0:
        return min(ds) + 1
    else:
        return "-1"
arr = list(map(int, input().split()))
k = int(input())
print(a(arr, k))
        
```

```python
def a(b):
    from decimal import Decimal
    c = 0
    for i in range(len(b)):
        if b[i] == ("+" or "-" or "/" or "*"):
            c = i
    print(c)
    d = Decimal(b[:c])
    f = Decimal(b[c+1:])
    print(d)
    print(f)
    if b[c] == "+":
        return round(d + f, 2)
    elif b[c] == "-":
        return round(d - f, 2)
    elif b[c] == "/":
        return round(d / f, 2)
    elif b[c] == "*":
        return round(d * f)
b = input()
print(a(b))
    
```

```python
def nhi_phan(n):
    ds = []
    while n != 0:
        ds.append(str(int(n % 2)))
        n //= 2
    ds = ds[::-1]
    b = int("".join(i for i in ds))
    return b
def e(x, y):
    total = 0
    c = nhi_phan(x)
    d = nhi_phan(y)
    f = len(str(c))
    g = len(str(d))
    if f == g:
        for i in range(f):
            if int(str(c)[i]) != int(str(d)[i]):
                total += 1
            else:
                continue
    elif f > g:
        total = f - g
        for i in range(g):
            if int(str(c)[i]) != int(str(d)[i]):
                total += 1
            else:
                continue
    elif f < g:
        total = g -f
        for i in range(f):
            if int(str(d)[i]) != int(str(c)[i]):
                total += 1
            else:
                continue
    return total
x, y = map(int, input().split())
print(e(x, y))
```

```python
def dang_cau(a, b):
    ds1 = []
    ds2 = []
    for i in range(len(a)):
        ds1.append([a[i], b[i]])
    for i in range(len(ds1)):
        for x in range(len(ds1)):
            if i != x:
                if ds1[i][0] == ds1[x][0]:
                    ds2.append([i, x])
                else:
                    continue
    total = 0
    for i in ds2:
        c = i
        d = c[::-1]
        if d in ds2:
            ds2.remove(d)
    for i in ds2:
        if ds1[i[0]][1] == ds1[i[1]][1]:
            total += 1
        else:
            continue
    if total == len(ds2):
        return True
    else:
        return False
a, b = map(str, input().split())
print(dang_cau(a, b))
```
